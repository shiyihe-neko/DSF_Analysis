```python
from typing import List, Any

def count_search_shortcuts(events: List[List[Any]], max_interval_ms: int = 2000) -> int:
    """
     windowEvents Meta/Control + F 
    - events:  [timestamp, eventType, key]
    - max_interval_ms: modifier  F 
    """
    count = 0
    last_mod_ts = None

    for ts, et, key in events:
        k = str(key).lower()
        e = str(et).lower()

        #  Meta/Ctrl 
        if k in ("meta", "control") and e.startswith("key"):
            last_mod_ts = ts

        #  F  modifier +1
        elif k == "f" and e.startswith("key") and last_mod_ts is not None:
            if ts - last_mod_ts <= max_interval_ms:
                count += 1
            # 
            last_mod_ts = None

    return count

#    #
# 1.  windowEvents 
#    events = test['answers']['writing-task-NL']['windowEvents']
# 2. 
#    print(count_search_shortcuts(events))
events = test['answers']['writing-task-NL_4']['windowEvents']
count_search_shortcuts(events)
```
